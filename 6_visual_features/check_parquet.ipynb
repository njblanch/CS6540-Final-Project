```python
import pandas as pd
```

```python
# df = pd.read_parquet('_6bNvy6IFAs.parquet')
df = pd.read_parquet('_6bNvy6IFAs.parquet')
# set video_id as str, all other as float
df['video_id'] = df['video_id'].astype(str)
print(df)
```

```python
audio_df = pd.read_csv("_6bNvy6IFAs.csv")
audio_df['video_number'] = audio_df['video_number'].astype(str)
# rename video_number to clip_num
audio_df = audio_df.rename(columns={"video_number": "clip_num"})
print(audio_df)
```

```python
# merge and see which rows are missing based of of clip_num and frame_number
merged = pd.merge(df, audio_df, on=["video_id", "clip_num", "frame_number"], how="outer", indicator=True)
# getitng only rows that are missing from either
missing = merged[merged['_merge'] == 'right_only']
print(missing)
```
