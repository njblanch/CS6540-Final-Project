```python
import matplotlib.pyplot as plt
import numpy as np

num = 16
# Define the file name
file_name = f"loss_{num}.txt"

# Initialize lists to store the parsed data
loss = []
primary_loss = []
reg_loss = []

# Parse the file
with open(file_name, "r") as file:
    for line in file:
        # Split the line into parts and extract relevant columns
        parts = line.strip().split(",")
        try:
            loss_value = float(parts[0].split(":")[1].strip())
            primary_loss_value = float(parts[1].split(":")[1].strip())
            reg_loss_value = float(parts[2].split(":")[1].strip())

            # Append to the lists
            loss.append(loss_value)
            primary_loss.append(primary_loss_value)
            reg_loss.append(reg_loss_value)
        except (IndexError, ValueError):
            # Skip lines that do not match the expected format
            continue

# Plot the data
time_steps = list(range(1, len(loss) + 1))

loss = np.sqrt(np.array(loss))
primary_loss = np.sqrt(np.array(primary_loss))

# Plot Loss
plt.figure(figsize=(6, 3))
plt.plot(time_steps, loss, label="Loss")
plt.xlabel("Time Step")
plt.ylabel("Loss")
plt.title("Loss Over Time")
plt.legend()
# plt.show()
plt.savefig(f"plots/loss_{num}.png")

# Plot Primary Loss
plt.figure(figsize=(6, 3))
plt.plot(time_steps, primary_loss, label="Primary Loss", color="orange")
plt.xlabel("Time Step")
plt.ylabel("Primary Loss")
plt.title("Primary Loss Over Time")
plt.legend()
# plt.show()
plt.savefig(f"plots/primary_loss_{num}.png")

# Plot Reg Loss
plt.figure(figsize=(6, 3))
plt.plot(time_steps, reg_loss, label="Reg Loss", color="green")
plt.xlabel("Time Step")
plt.ylabel("Reg Loss")
plt.title("Reg Loss Over Time")
plt.legend()
# plt.show()
plt.savefig(f"plots/reg_loss_{num}.png")
```
